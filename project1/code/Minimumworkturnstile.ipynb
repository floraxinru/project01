```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('june82019.txt')
```

```python
df['DATETIME'] = df['DATE'] + ' ' + df['TIME']
```

```python
df['DATETIME'] = pd.to_datetime(df['DATETIME'])
```

```python
df_stations = df.loc[df['STATION'].apply(lambda x: x in ['59 ST'])]
df_stations2 = df_stations.loc[df_stations['SCP'].apply(lambda x: x in ['02-00-00'])]
df_stations2
```

```python
x = df_stations2['DATETIME']
y = df_stations2['ENTRIES']
plt.plot(x,y)
```
