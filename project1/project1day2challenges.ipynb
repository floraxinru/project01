```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# You can configure the format of the images: ‘png’, ‘retina’, ‘jpeg’, ‘svg’, ‘pdf’.
%config InlineBackend.figure_format = 'svg'
# this statement allows the visuals to render within your Jupyter Notebook
%matplotlib inline
```

```python
data_x = np.arange(7) #days of the week for one station over multiple weeks (rainbow band)
data_y = np.random.randint(500,1000,7) # random integer number between 0 and 9 (10 is excluded)
#first week
plt.plot(data_x,data_y);
```

```python
#second week
data_x1 = np.arange(7) #days of the week for one station over multiple weeks (rainbow band)
data_y1 = np.random.randint(500,1000,7) # random integer number between 0 and 9 (10 is excluded)
#third week
data_x2 = np.arange(7) #days of the week for one station over multiple weeks (rainbow band)
data_y2 = np.random.randint(500,1000,7) # random integer number between 0 and 9 (10 is excluded)
```

```python
plt.plot(data_x,data_y);
#plt.grid()
ticks_x = np.linspace(0, 7, 8) #8 tick marks between 0 and 7, counting 0 as one tick?
station_names = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
plt.xticks(ticks_x, station_names, fontsize = '10', family = 'sans-serif', rotation=40, color='orange');
plt.title('Weekly Commute Numbers',fontsize = 25,style='italic',weight='bold', color='blue');
plt.xlabel('Days of the Week', fontsize = 15, weight = 'bold', family = 'sans-serif', color='black')
plt.ylabel('Daily Count',fontsize = 15, weight = 'bold', family = 'sans-serif',color='black');

plt.plot(data_x1,data_y1);
plt.plot(data_x2,data_y2);

plt.legend(['Week1','Week2','Week3'],shadow = True, loc = 0);
```
